```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=[34,56,90,200,300]
x=['Python','English','Hindi','AI','DA']

plt.pie(y,labels=x)
plt.legend()
plt.show()
```

```python
sales = [34,89,23,56,33,34,12,90]
month = ['Jan','Feb','Mar','April','May','Sept','Jan','Mar']
profit = [8,10,56,50,23,33,90,12]
loss = [23,45,89,45,23,1,2,3]
plt.pie(profit,labels=month, autopct='%1.f%%',radius=2,pctdistance=0.5 , labeldistance=1.1)
plt.show()
```

```python
sec_A=np.array([23,45,12,19,54])
subject=["Science","AI","PYthon","English","DBMS"]
data = [0, 0.5, 0, 0,0]
plt.pie(sec_A,labels=subject,explode=data,shadow = True)
plt.legend(title="subject",fontsize=9,loc="upper right")
plt.show()
```
