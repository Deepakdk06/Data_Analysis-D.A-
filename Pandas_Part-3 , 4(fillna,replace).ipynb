# fillna

```python
data.fillna(6) # will be fill '6' on every NaN value
```

```python
data
```

```python
data.fillna({'name':'raj'}) # to fill colunm wise NaN value in categorical data
```

```python
data.fillna({'age':24.5})  # to fill colunm wise NaN value in numarical data
```

```python
data
```

```python
data.fillna(method='backfill')  #there will be fill NaN value as mentioned value after NaN
```

```python
data.fillna(method='bfill')  #same as backfill
```

```python
data
```

```python
data.fillna(method='pad') #apposit of backfill and bfill
```

```python
data.fillna(method='ffill') # same as pad
```

```python
data
```

```python
data.fillna(value=10,limit=1)  #Only replace the first NaN element from the colunm if use limit=1.
```

```python
data
```

```python
data.fillna({'name': 'ab', 'age': 25, 'marks': 67})
```

```python
values = {'name': 'ab', 'age': 25, 'marks': 67}
data.fillna(value=values)
```
